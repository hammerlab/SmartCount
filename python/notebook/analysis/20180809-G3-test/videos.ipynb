## Video Generation

This notebook contains code to generate video files for G3 chip apartments.

```python
%run -m celldom.nb.logging
import os
import os.path as osp
import numpy as np
import pandas as pd
import celldom
import glob
from celldom.config import experiment_config
from celldom.execute import visualization
```

```python
# Choose an experiment configuration and directory in which to store video data
exp_config = experiment_config.get_exp_config_by_name('experiment_example_G3')
output_dir = osp.join(celldom.get_data_dir(), 'output', 'tests', 'videos_02')
output_dir
```

```python
# Show all the sample raw files
files = glob.glob('/lab/data/celldom/dataset/dataset04/20180614Orange1_0_1uM000hr/*.tif')
files
```

```python
# Run this function to generate videos 
visualization.generate_apartment_videos(exp_config, files, output_dir, video_type='gif')
```
