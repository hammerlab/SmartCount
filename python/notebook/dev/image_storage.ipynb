```python
def save_image(key, image):
    store.put('/shape/' + key, pd.Series(img.shape))
    store.put('/data/' + key, pd.Series(img.ravel()))
    
def load_image(key):
    shape = tuple(store.get('/shape/' + key))
    data = store.get('/data/' + key)
    return data.values.reshape(shape)
```
