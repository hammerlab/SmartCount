## Video Generation Example (Simple Version)

Videos of apartments can also be generated this way, where all that is necessary is:

1. A directory to save data in
2. A list of raw, multi-apartment images to process

From there, the modules used here would extract individual apartment images.

```python
%run -m celldom.nb.logging
import os
import os.path as osp
import numpy as np
import pandas as pd
import celldom
import glob
from celldom.config import experiment_config
from celldom.execute import visualization
```

```python
# Choose an experiment configuration and directory in which to store video data
exp_config = experiment_config.get_exp_config_by_name('experiment_example_01')
output_dir = osp.join(celldom.get_data_dir(), 'output', 'tests', 'videos_01')
output_dir
```

```python
# Choose a bunch of raw files where in this case, we want everything near
# apartment 016 and street 024 and since there are multiple apartments in each,
# the results will ultimately include all apartments in the vicinity
files = glob.glob('/lab/data/celldom/dataset/dataset03/*NoDrug*/BF_ST_024_APT_016_*.tif')
[f[-69:] for f in files]
```

```python
# Run this function to generated videos 
visualization.generate_apartment_videos(exp_config, files, output_dir, video_type='gif')
```
